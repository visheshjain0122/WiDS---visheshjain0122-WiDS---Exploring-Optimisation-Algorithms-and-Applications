# Exercise 1 Part 1

the given function

$f(x_1,x_2) = x_1^2 + 4x_1x_2 + 1600x_2^2$

The minimiser is

$(x_1,x_2) = (0,0)$

And the minimum is $0$

```python
#Exercise 1 part 1

import numpy as np

def hess(x,y):
    h = np.zeros((2,2))
    h[0][0] = 2
    h[0][1] = 4
    h[1][0] = 4
    h[1][1] = 3200
    return h

def hess_inverse(x,y):
    h = hess(x,y)
    h_inv = np.linalg.inv(h)
    return h_inv

def norm_inf(a):
    #we pass a 2x2 matrix 'a' as a parameter
    row_sum_1 = abs(a[0][0]) + abs(a[0][1])
    row_sum_2 = abs(a[1][0]) + abs(a[1][1])

    if (row_sum_1 > row_sum_2):
        inf_norm = row_sum_1

    else:
        inf_norm = row_sum_2

    return inf_norm


#now it is know to us that the condition number is
# = ||A||_inf * ||A^-1||_inf
# So,

#condition_number = norm_inf(hess(x,y)) * norm_inf(hess_inverse(x,y))

#here x,y haven't been defined yet
#however, we can check whether the condition number is too high at
#any particular point in a given range of x,y using the above code

ill_count = 0

for i in range(-4000,4001):
    for j in range(-1,2):
        condition_number = norm_inf(hess(j,i)) * norm_inf(hess_inverse(j,i))

        if(condition_number > 1):
          print("Ill-condtioned at")
          print("x = ", j)
          print("y = ", i)
          print()

          ill_count = ill_count + 1


print("Total points at which hessian is ill-conditioned = ", ill_count)
```

# Exercise 2 Part 2a

A matrix $D$ is said to be symmetric positive definite if and only if for all vectors

$x^TDx >0$ for all column matrices $x$ not equal to zero in $R^n$

In steepest descent, when we say

$x` = x - nDgrad(f)$

Basically the direction of our descent must always be negative

Which implies that if we take descent direction

$-grad(f)^TDgrad(f) < 0$

(must always be less than zero for descent)

This implies that

$grad(f)^TDgrad(f) > 0$

Which proves that $D$ is symmetric and positive definite

# Exercise 2 Part 2b

The matrix $Q$ whose condition number we wish to find is nothing but

$Q = hess_y(g(y)) = M^T.hess_x(f(x)).M$

where

$D = M.M^T$

Where $M$ is a transformation on $x$

Such that

$y = M.x$

A smaller condition value for this matrix implies a better convergence rate.

Based on the matrix $Q$, we can make matrix $D$ (which is diagonal) such that it is similar to the inverse of the hessian but not entirely the same.

We have been told to take $D$ as a diagonal matrix.

However, the thing is with diagonal matrices is that they are always ill-conditioned or the condition number will always be greater than $1$.

Regardless if we multiply a diagonal matrix by an scaler, it will always have condition number more than $1$

Back to our consideration for $D$.

If we were to make a diagonal matrix out of $D$, we can construct it similar to the hessian such that

$D(i,i) = max(ϵ, d^2f(x_k)/d^2x_i)^{-1}$ where $ϵ>0 $

Also, this $D$ is positive definite as each diagonal element will always be greater than or equal to $1/ϵ$ which is always positive definite and symmetric.

Using our code, we can also check the number of points within our range at which $D$ is ill-conditioned.

```python
#Exercise 3 part 3 - gradient descent with/without scaling using backtracking for learning rate

import numpy as np
import math as m
import matplotlib.pyplot as plt

#definig our function
def func(x):
    return x[0]**2 + 4*x[0]*x[1] + 1600*x[1]**2

#defining our gradient
def grad(x):
    g = np.zeros((2,1))
    g[0][0] = 2*x[0] + 4*x[1]
    g[1][0] =  3200*x[1] + 4*x[0]

    return g

#defining our matrix D
def D(x):
    h = np.zeros((2,2))
    h[0][0] = 1/2
    h[0][1] = 0
    h[1][0] = 0
    h[1][1] = 1/3200

    return h

def hess(x):
    h = np.zeros((2,2))
    h[0][0] = 2
    h[0][1] = 4
    h[1][0] = 4
    h[1][1] = 3200
    return h


#defining the norm calculation for the gradient

def norm(gradient): #here we are taking the gradient as an array type argument for our function

    #sum = 0

    #for i in range(2):
      #sum = sum + (gradient[i][0])**2

    #return m.sqrt(sum)
    return np.linalg.norm(gradient)


#now we shall define our descent with scaling algorithm

def descent_with_scaling(x0, toler, p0):

    #starting with the given value of x initial
    x = x0

    #initialising some parameters
    g = grad(x)
    #h = hess(x)
    D_mat = D(x)
    L2 = norm(g)
    count = 0

    while(L2 > toler):

       #initialising line search parameters

        a0 = 1
        p = p0
        gamma = 0.5

        #initialising the direction parameter

        d =  - D_mat @ g

       #now the line search loop

        while (func(x + a0*d) > func(x) + gamma*a0*g.T@ d): # Check the RHS once, whether it is supposed to be g.T @ d or d.T @ d
            a0 = a0 * p


        #updating values of x
        x = x + a0*d

        #initialising new gradient and norm wrt to new x
        g = grad(x)
        L2 = norm(g)

        #and updating the count
        count = count + 1

    return count, x, func(x)





#now we shall define our descent without scale algorithm

def descent_no_scaling(x0, toler, p0):

    #starting with the given value of x initial
    x = x0

    #initialising some parameters
    g = grad(x)
    #h = hess(x)
    #D_mat = D(x)
    L2 = norm(g)
    count = 0

    while(L2 > toler):

       #initialising line search parameters

        a0 = 1
        p = p0
        gamma = 0.5

        #initialising the direction parameter

        d =  - g

       #now the line search loop

        while (func(x + a0*d) > func(x) + gamma*a0*g.T@ d): # Check the RHS once, whether it is supposed to be g.T @ d or d.T @ d
            a0 = a0 * p


        #updating values of x
        x = x + a0*d

        #initialising new gradient and norm wrt to new x
        g = grad(x)
        L2 = norm(g)

        #and updating the count
        count = count + 1


    return count, x, func(x)



#now the portion to initialise everything and run the final code

toler = 0.0000000000001 #10^-11

#initial value
x0 = np.array([[4000], [1]])

#defining the start points for the backtracking algorithm
p0 = np.array([0.9, 0.8, 0.75, 0.6, 0.5, 0.4, 0.25, 0.1, 0.01])

#also creating list to append values to plot iterations vs a0
P = []
I_wo = [] #without scaling
I_w = [] #with scaling


for i in range(len(p0)):
    p_init = p0[i]
    P.append(p_init)


    iter_w, x_w, f_opt_w = descent_with_scaling(x0, toler, p_init)
    print("For gradient descent with scaling")
    print("For p value = ", p_init)
    print()
    print("iterations = ", iter_w)
    print()
    print("minimiser = ", x_w)
    print()
    print("minimum = ", f_opt_w)

    I_w.append(iter_w)


    iter_wo, x_wo, f_opt_wo = descent_no_scaling(x0, toler, p_init)
    print("For gradient descent without scaling")
    print("For p value = ", p_init)
    print()
    print("iterations = ", iter_wo)
    print()
    print("minimiser = ", x_wo)
    print()
    print("minimum = ", f_opt_wo)

    I_wo.append(iter_wo)


plt.plot(P, I_wo)
plt.plot(P,I_w)
plt.xlabel('Initial value for p')
plt.ylabel('Total Iterations')
plt.legend(["Without Scaling", "With Scaling"])

plt.show()
```
