```python
# Exercise 2 part 1 algorithm

import numpy as np

#defining our function
def objfunc(x):
  return np.sqrt(x**2 + 5)

#defining an algorithm that takes values of e, L and A to produce our grid
def algo1(e, L, A):
  grid_resolution = 2 * e / L
  X = np.arange(-A/10 - 2, A/10 + 2, grid_resolution)
  fn = [objfunc(x) for x in X]

  xtilda = X[np.argmin(fn)]
  return xtilda

A = 7
L = 1
epsilons = []
for i in range (1, 9):
  epsilons.append(10 ** -i)

results = []

# here we carry out the algorithm with different values of epsilon

for e in epsilons:
  xtilda = algo1(e, L, A)
  results.append([e, xtilda])

print(results)
```

It appears that with the smaller the value of epsilon the closer and closer our x value gets to our min 0. Its as if it **oscillates** around our desired minimum value.

```python
# Exercise 2 part 2 algorithm

import numpy as np

# defining our function
def objfunc(x):
  return np.sqrt(x[0]**2 + x[1]**2)

#defining an algorithm that takes values of e, L, A to produce our grid
def algo1(e, L, A):
  grid_resolution = 2 * e / L
  X0 = np.arange(-A/10 - 2, A/10 + 2, grid_resolution)
  X1 = np.arange(-A/5 - 2, A/5 + 2, grid_resolution)
  X0_mesh, X1_mesh = np.meshgrid(X0, X1)
  X = np.column_stack((X0_mesh.ravel(), X1_mesh.ravel()))
  fn = [objfunc(x) for x in X]

  xtilda = X[np.argmin(fn)]
  return xtilda

A = 2
L = 1
epsilons1 = []
epsilons2=[]
epsilons3=[]


for i in range (1, 4):
  epsilons1.append(10 ** -i)
  epsilons2.append(2* 10**-i)
  epsilons3.append(5* 10**-i)

results = []

#here we carry out the algorithm with different values of epsilon

for e in epsilons1:
  xtilda = algo1(e, L, A)
  results.append([e, xtilda])

for e in epsilons2:
  xtilda = algo1(e, L, A)
  results.append([e, xtilda])

for e in epsilons3:
  xtilda = algo1(e, L, A)
  results.append([e, xtilda])

print(results)
```

Similarly overhere, we end up **oscillating** around our minima, which is

$(x,y) = (0,0)$
