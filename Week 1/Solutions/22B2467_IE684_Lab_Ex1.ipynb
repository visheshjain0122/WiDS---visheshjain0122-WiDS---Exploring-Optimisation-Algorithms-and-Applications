```python
# Exercise 1 Part 1

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def f(x, y):
    return 100*x**4 + 100*y**2 - 200*y*(x**2) + x**2 - 2*x + 1

# Create a grid of points in x, y, z directions
x = np.linspace(-4, 4, 100)
y = np.linspace(-4, 4, 100)

[X, Y]= np.meshgrid(x, y)

# Calculate the values of the function on the grid
F = f(X, Y)

fig, ax = plt.subplots(1, 1)

#making the contour features
ax.contour(X, Y, F)

ax.set_title('Contour Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')

#show the plot
plt.show()
```
